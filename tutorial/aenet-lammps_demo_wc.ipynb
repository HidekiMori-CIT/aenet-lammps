# 4. Compiling ænet as LAMMPS library
Patch and compile aenet library. Please check the below two files are created: 
1. libaenet.a: library of aenet
2. Makefile.lammps: linker of aenet and lammps 

```python
!patch -u -p1 -d aenet/ < aenet-2.0.4_lammps.patch
%cd aenet/src/
!make -f makefiles/Makefile.gfortran_serial lib
%cd ../../
```

# 5. Compiling LAMMPS with ænet library

```python
%cd ../../src/
!make yes-user-aenet
!make mpi
```

#6. How to run LAMMPS with ANN potential
In ANN/Fe/ directory parameter file and example input file for LAMMPS
1. Fe.10tw-10tw.ann: parameter file of ANN potential for BCC iron
2. in.aenet_mm: example input file for LAMMPS, lattice optimization of BCC iron
3. in.aenet_nve: example input file for LAMMPS, run nve MD of BCC iron

To check the LAMMPS work properly, for example, in ANN/Fe/ directory

```python
%cd ../../aenet-lammps-2021May04/ANN/Fe/
!../../../lammps-stable_29Oct2020/src/lmp_mpi -i in.aenet_mm
```

#7. Checking the result
After job done, please check lattice constant (lx) in log.lammps and log.lammps.g++_mm

In in.aenet_mm the following two lines activate the ANN potential:

> pair_style      aenet

> pair_coeff      * * v01 Fe 10tw-10tw.ann Fe

In pair_coeff line, **v01** means to use our (Mori and Ozaki) original version. If you chose **v00**, you can use parameter file from original aenet package. The Fe between v01 and 10tw-10tw.ann set element(s). The term:10tw-10tw.ann is **file mask** for name of parameter file, in this case, file name is set as Fe.10tw-10tw.ann. Please see TiO2 directory for binary system.
